## Import Library

```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
```

## Import CSV

```python
df = pd.read_csv("/content/drive/MyDrive/online_retail_II.csv", header = 0)
df = df[df['Price'].between(0, 100, inclusive='right')].reset_index(drop=True)
df
```

```python
df.info()
```

## Membuat Data Binning

Membuat bins (dengan banyaknya bins tertentu dan masing masing panjang nya)

```python
df1 = df.copy()
category = ['Cheap', 'Medium', 'Expensive']
df1['bin1'] = pd.cut(df1['Price'], 3, labels=category, include_lowest=True)
df1
```

```python
binning1_summary = df1.groupby('bin1').agg(row_cnt=('StockCode', 'count'), min_price=('Price', 'min'), max_price=('Price', 'max'))
binning1_summary['bin_interval'] = binning1_summary['max_price'] - binning1_summary['min_price']
binning1_summary
```

```python
df2 = df1.copy()
bins = np.linspace(min(df['Price']), max(df['Price']), 4)
category = ['Cheap', 'Medium', 'Expensive']
df2['bin2'] = pd.cut(df2['Price'], bins, labels=category, include_lowest=True)
df2
```

```python
bins
```

```python
binning2 = df2.groupby('bin2').agg(row_cnt=('StockCode', 'count'), min_price=('Price', 'min'), max_price=('Price', 'max'))
binning2['bin2_interval'] = binning2['max_price'] - binning2['min_price']
binning2
```

## Binning dengan panjang masing-masing interval tertentu

```python
df3 = df2.copy()
interval_range = pd.interval_range(start=min(df['Price']), freq = 10, end=max(df['Price']))
df3['bin3'] = pd.cut(df3['Price'], bins=interval_range, include_lowest=True)
df3
```

```python
interval_range
```

```python
binning3 = df3.groupby('bin3').agg(row_cnt=('StockCode', 'count'), min_price=('Price', 'min'), max_price=('Price', 'max'))
binning3['bin3_interval'] = binning3['max_price'] - binning3['min_price']
binning3
```

## Binning dengan panjang masing masing bin yang ditentukan oleh kebutuhan sendiri

```python
df4 = df3.copy()
bins_interval = [0, 30, 80, 100]
category = ['Cheap', 'Medium', 'Expensive']
df4['df4_bins'] = pd.cut(df4['Price'], bins=bins_interval, labels=category, include_lowest=True)
df4
```

```python
bin4_summarry = df4.groupby('df4_bins').agg(row_cnt=('StockCode', 'count'), min_price=('Price', 'min'), max_price=('Price', 'max'))
bin4_summarry['bin4_group'] = bin4_summarry['max_price'] - bin4_summarry['min_price']
bin4_summarry
```

## Binning dengan banyaknya bins tertentu dan banyaknya anggota setiap bin sama

```python
df5 = df4.copy()
category = ['Cheap', 'Medium', 'Expensive']
df5['bin5'] = pd.qcut(df5['Price'], 3, labels=category)
df5
```

```python
summary5 = df5.groupby('bin5').agg(row_cnt=('StockCode', 'count'), min_price=('Price', 'min'), max_price=('Price', 'max'))
summary5['interval5_summary'] = summary5['max_price'] - summary5['min_price']
summary5
```

#  Mengelompokkan data dengan CASE STATEMENT

```python
df6 = df5.copy()
```

```python
df6['price_class'] = np.select(
    [
    (df['Description'].str.lower().str.contains('sweetheart')) & (df['Price'] <= 3),
    (df['Description'].str.lower().str.contains('sweetheart')) & (df['Price'].between(3, 7, inclusive='right')),
    (df['Description'].str.lower().str.contains('sweetheart')) & (df['Price'] >= 8),
    (df['Description'].str.lower().str.contains('spotty')) & (df['Price'] <= 2),
    (df['Description'].str.lower().str.contains('spotty')) & (df['Price'].between(3, 6, inclusive='right')),
    (df['Description'].str.lower().str.contains('spotty')) & (df['Price'] >= 6 ),
    (df['Description'].str.lower().str.contains('french')) & (df['Price'] <= 2),
    (df['Description'].str.lower().str.contains('french')) & (df['Price'].between(3, 5, inclusive='right')),
    (df['Description'].str.lower().str.contains('french')) & (df['Price'] >= 5)
    ],
    [
        'Sweetheart Cheap',
        'Sweetheart Medium',
        'Sweetheart Expensive',
        'Spotty Cheap',
        'Spotty Medium',
        'Spotty Expensive',
        'French Cheap',
        'French Medium',
        'French Expensive'
    ],
    default='Unknown'
)
df6
```

```python
summarry_class = df6.groupby('price_class').agg(row=('StockCode', 'count'), min_price=('Price', 'min'), max_price=('Price', 'max'))
summarry_class['inteerval_class'] = summarry_class['max_price'] - summarry_class['min_price']
summarry_class
```
